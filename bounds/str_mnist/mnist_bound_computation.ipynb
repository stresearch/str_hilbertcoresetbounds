```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# STR Mnist Bounds
We compute the generalization bounds for the STR Mnist model.
We load the results of the coreset calculation script and calculate how the bound scales with the number of iterations.
Then we compile the results from the checkpoints to compute the final bound.
## Checkpoint 500

```python
res = np.load("../../bound_calculation/hilbert_coreset_results/str_coreset_mnist_results/algorithmic_bound/results_500.npz")
iters = res["iters"]
core_sizes = res["coresize"]
gen_err_500 = res["genErr"]
w_minus_p_norm = res["wminuspnorm_full"]
l_norm = res["lnorm_prior"]
l_lw_norm = res["llwnorm"]
lw_norm = res["lwnorm"]
w_norm = res["wnorm"]

params = np.load("../../bound_calculation/hilbert_coreset_results/str_coreset_mnist_results/exponential_bound/constants_500.npz")
sigma = params["sigma"]
xi = params["xi"]
eta = params["eta"]
eta_bar = params["eta_bar"]
beta = params["beta"]

correct_diff = w_norm**2 * xi * math.sqrt(2 * math.log(2 * 70000**2 / 0.05) / 7000)
correct_emp = w_minus_p_norm**2 * xi * math.sqrt(2 * math.log(2 * 70000**2 / 0.05) / 7000)
print(correct_diff)
print(correct_emp)

top = sigma * eta * eta_bar * beta
first = eta_bar**2 * beta**(-2 * (iters-1))
second = eta**2 * (iters-1)
exponential_bound = top / np.sqrt(first + second)

alg_bound_500 = np.sqrt(lw_norm**2 + correct_emp) + np.sqrt(l_lw_norm**2 + correct_diff)
exp_bound_500 = np.sqrt(lw_norm ** 2 + correct_emp) + np.sqrt(exponential_bound ** 2 + correct_diff)
plt.plot(iters, alg_bound_500, color='b', label='Algorithmic Bound')
plt.plot(iters, exp_bound_500, color='r', label='Exponential Bound')
plt.plot(iters, gen_err_500, color='g', label="Expected Risk")
plt.title("Algorithmic bound for 500 sample size")
plt.xlabel("Number of FW iterations")
plt.legend();
```

## Overall Bound

```python
train_size = np.array([500, 3000, 10000, 30000])
alg_bound = np.zeros(4)
exp_bound = np.zeros(4)
exp_risk = np.zeros(4)

alg_bound[0] = alg_bound_500[-1]
alg_bound[1] = alg_bound_3000[-1]
alg_bound[2] = alg_bound_10000[-1]
alg_bound[3] = alg_bound_30000[-1]

exp_bound[0] = exp_bound_500[-1]
exp_bound[1] = exp_bound_3000[-1]
exp_bound[2] = exp_bound_10000[-1]
exp_bound[3] = exp_bound_30000[-1]

exp_risk[0] = gen_err_500[-1]
exp_risk[1] = gen_err_3000[-1]
exp_risk[2] = gen_err_10000[-1]
exp_risk[3] = gen_err_30000[-1]

print("Exponential Bound: ", exp_bound)
print("Algorithmic Bound: ", alg_bound)
print("Expected Risk: ", exp_risk)

ax = plt.axes(xscale='linear', yscale='linear')
ax.grid();
ax.plot(train_size, alg_bound,  color='blue', marker='x', linestyle='solid',linewidth=1, markersize=8, label="Coreset Bound I")
ax.plot(train_size, exp_bound, color='red', marker='^', linestyle='solid',linewidth=1, markersize=8,label="Coreset Bound II")
ax.plot(train_size, exp_risk, color='green', marker='o', linestyle='dashed',linewidth=1, markersize=8,label=" Expected Empirical Risk ")
ax.set_xlabel('Training set size (S)', fontsize= 13, fontweight='bold')
ax.set_ylabel('Risk Bound/EER', fontsize= 13, fontweight='bold')
ax.set_title('Coreset Bound -- Binary Mnist Dataset', fontsize= 13, fontweight='bold')
ax.tick_params(labelsize = 'large')
#ax.set_ylim([0.1, 3])
ax.legend();
```
