```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# STR Mnist Bounds
We compute the generalization bounds for the STR Mnist model.
We load the results of the coreset calculation script and calculate how the bound scales with the number of iterations.
Then we compile the results from the checkpoints to compute the final bound.
## Checkpoint 500

```python
res = np.load("../../bound_calculation/hilbert_coreset_results/str_coreset_mnist_results/algorithmic_bound/results_500.npz")
iters = res["iters"]
core_sizes = res["coresize"]
gen_err_500 = res["genErr"]
w_minus_p_norm = res["wminuspnorm_full"]
l_norm = res["lnorm_prior"]
l_lw_norm = res["llwnorm"]
lw_norm = res["lwnorm"]
w_norm = res["wnorm"]

params = np.load("../../bound_calculation/hilbert_coreset_results/str_coreset_mnist_results/exponential_bound/constants_500.npz")
sigma = params["sigma"]
xi = params["xi"]
eta = params["eta"]
eta_bar = params["eta_bar"]
beta = params["beta"]

correct_diff = w_norm**2 * xi * math.sqrt(2 * math.log(2 * 70000**2 / 0.05) / 7000)
correct_emp = w_minus_p_norm**2 * xi * math.sqrt(2 * math.log(2 * 70000**2 / 0.05) / 7000)
print(correct_diff)
print(correct_emp)

top = sigma * eta * eta_bar * beta
first = eta_bar**2 * beta**(-2 * (iters-1))
second = eta**2 * (iters-1)
exponential_bound = top / np.sqrt(first + second)

alg_bound_500 = lw_norm + l_lw_norm + correct_diff + correct_emp
exp_bound_500 = lw_norm + exponential_bound + correct_diff + correct_emp
plt.plot(iters, alg_bound_500, color='r', label='Algorithmic Bound')
plt.plot(iters, exp_bound_500, color='b', label='Exponential Bound')
plt.plot(iters, gen_err_500, color='g', label="Expected Risk")
plt.title("Algorithmic bound for 500 sample size")
plt.xlabel("Number of FW iterations")
plt.legend();
```
